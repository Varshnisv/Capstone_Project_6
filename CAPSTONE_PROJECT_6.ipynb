## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer

Hypothesis Statement 1: Yes Bank's average closing stock prices before and after 2018 differ significantly.

The second hypothesis states that, in comparison to the pre-2018 period, the monthly distribution of closing prices exhibits noticeably greater volatility (higher variance) in the post-2018 period.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer

Null Hypothesis (H0): The average closing prices before and after 2018 do not differ significantly.

Alternative Hypothesis (H1): The average closing prices prior to and following 2018 differ significantly.




#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import ttest_ind

# Perform t-test for independent samples (Pre-2018 vs Post-2018 closing prices)
pre_close_2018 = pre_2018['Close']
post_close_2018 = post_2018['Close']

# Perform the t-test
t_stat, p_value = ttest_ind(pre_close_2018, post_close_2018)

# Output the results
print(f"T-Statistic: {t_stat}")
print(f"P-Value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("Rejected Null Hypothesis")
else:
    print("Failed to Reject Null Hypothesis")
````

##### Which statistical test have you done to obtain P-Value?

Answer
The Independent Samples T-Test, sometimes referred to as the Two-Sample T-Test, is the statistical test that is employed; it is specifically implemented using scipy.stats.ttest_ind.


##### Why did you choose the specific statistical test?

Answer
Examine the means of two separate groups (closing prices before and after 2018).
Assess whether the two groups differ significantly from one another.
- Examine the null hypothesis, which states that the means of the two groups are equal.

The test is applied in this situation to:

Examine the effects of an event (perhaps a change in policy, a change in the market, or another noteworthy event in 2018).
Assess whether there is a substantial difference between the closing prices before and after the incident.

The following are the alternative and null hypotheses:

The null hypothesis (H0) states that the means of the closing prices before and after 2018 are equal (μ1 = μ2).
- Alternative Hypothesis (H1): μ1 ≠ μ2 (the means of closing prices before and after 2018 differ).

We can ascertain the likelihood of detecting the test results under the null hypothesis by computing the p-value.


### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer

Null Hypothesis (H0): The variance (volatility) of closing prices prior to and following 2018 does not differ significantly.

Alternative Hypothesis (H1): After 2018, closing prices have a substantially greater variance (volatility) than they did before 2018.


#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

from scipy.stats import f

# Calculate the variances for both periods
before_2018_variance = pre_2018['Close'].var()
after_2018_variance = post_2018['Close'].var()

# Perform the F-test
f_stat = before_2018_variance / after_2018_variance
dfn = len(pre_2018) - 1
dfd = len(post_2018) - 1
# Calculate the critical value at alpha = 0.05
alpha = 0.05
f_critical = f.ppf(1 - alpha, dfn, dfd)

# Output the results
print(f"F-Statistic: {f_stat}")
print(f"Critical Value: {f_critical}")

# Interpretation
if f_stat > f_critical:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis.")
````

##### Which statistical test have you done to obtain P-Value?

Answer
The F-Test for equality of variances, which is specifically implemented using scipy.stats.f, is the statistical test that is employed.


##### Why did you choose the specific statistical test?

Answer
Using the F-Test, one can:

Examine the differences (volatility) between two separate groups (closing prices before and after 2018).
Assess whether the volatility during the two periods differs significantly.
Examine the null hypothesis, which states that the variances of the two groups are equal.

The test is applied in this situation to:

Examine how an event—likely a change in policy, a movement in the market, or another noteworthy event in 2018—affects market volatility.
Check to see if there is a noticeable difference in volatility before and after the incident.

The following are the alternative and null hypotheses:

σ1² = σ2² (pre-2018 and post-2018 variances are equal) is the null hypothesis (H0).
σ1² ≠ σ2² (pre-2018 and post-2018 variances diverge) is the alternative hypothesis (H1).



### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer

Null Hypothesis (H0): After 2018, there has been no discernible upward or falling trend in stock prices.

Alternative Hypothesis (H1): The impact of the 2018 fraud case is reflected in a notable higher or negative trend in stock prices after 2018.




#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import mannwhitneyu

# Mann-Kendall Test (simplified version for trend detection)

post_2018['Trend'] = post_2018['Close'].diff().apply(lambda x: 1 if x > 0 else (-1 if x < 0 else 0))

# Count the positive and negative trends
positive_trends = post_2018['Trend'].sum()
negative_trends = len(post_2018) - positive_trends
# For simplification, using a non-parametric approach to test the trend (note this isn't a formal Mann-Kendall test, more of a trend analysis)
u_stat, p_value = mannwhitneyu([positive_trends], [negative_trends], alternative='two-sided')

# Output results
print(f"U-Statistic: {u_stat}")
print(f"P-Value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
````

##### Which statistical test have you done to obtain P-Value?

Answer
The Mann-Whitney U Test, a non-parametric test built with scipy.stats.mannwhitneyu, is the statistical test that is employed. It should be noted that this is not a formal Mann-Kendall test, but rather a simplified trend analysis.



##### Why did you choose the specific statistical test?

Answer
The purpose of the Mann-Whitney U Test is to:

Look for patterns in the closing prices after 2018.
Compare the quantity of price changes (trends) that are positive and negative.
Analyse the data to see if there is a noteworthy increase or declining trend.

The test is applied in this situation to:

Examine the importance and direction of price changes.
Determine any possible patterns or trends.

The following are the alternative and null hypotheses:

There is no discernible trend (equal numbers of positive and negative trends), according to the null hypothesis (H0).
Significant trend (unequal number of positive and negative trends) is the alternative hypothesis (H1).

We can ascertain whether a significant trend exists by computing the p-value and U-statistic. The null hypothesis is rejected if the p-value is less than 0.05, which denotes a significant trend.


## ***6. Feature Engineering & Data Pre-processing***

### 1. Handling Missing Values

````python
# Handling Missing Values & Missing Value Imputation

mv= df1.isnull().sum()
print(mv)
df1['Open'] = df1['Open'].fillna(df1['Open'].mean())  # Mean Imputation
df1['High'] = df1['High'].fillna(df1['High'].median())  # Median Imputation
df1['Low'] = df1['Low'].fillna(df1['Low'].mean())  # Mean Imputation
df1['Close'] = df1['Close'].fillna(df1['Close'].median())  # Median Imputation
df1['Date'] = df1['Date'].fillna(method='ffill')

````

#### What all missing value imputation techniques have you used and why did you use those techniques?

Answer

Managing missing values is one of the most crucial parts of preparing the data for analysis. Lack of data could lead to skewed analysis or poorer model performance. Here's how to handle values that are missing:

Locate Missing Values: To begin, search for missing values using the isnull() or isna() methods. The columns and quantity of missing data can be summarised in this way. Imputation: There are several ways to handle missing values once they have been identified. The approach used depends on the kind of data and the quantity of missing data.

Mean/Median Imputation (for Numerical Columns):
When to Use It: This approach is effective when the missing data is in numerical columns and you don't want to lose a lot of records. Why? Imputing using the mean (for less skewed data) or median (for skewed data) preserves the overall distribution and lessens the impact on the analysis.

Mode Imputation (For Categorical Columns):

When to Apply It: For categorical characteristics, the most popular category is a great technique to fill in missing values. Why: This method maintains categorical data consistency and safeguards important records.



### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
# IQR for 'Close' column
Q1 = df1['Close'].quantile(0.25)
Q3 = df1['Close'].quantile(0.75)
IQR = Q3 - Q1
lb = Q1 - 1.5 * IQR #Lower Bound
ub = Q3 + 1.5 * IQR #Upper Bound

# Find and remove outliers
outliers = df1[(df1['Close'] < lb) | (df1['Close'] > ub)]
print("Total of outliers in 'Close':", len(outliers))

# Option 1: Removing outliers
df1 = df1[(df1['Close'] >= lb) & (df1['Close'] <= ub)]

# Option 2: Caping the outliers
df1['Close'] = df1['Close'].clip(lower=lb, upper=ub)
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

Answer
To reduce the impact of extreme values without eliminating data points, capping is used to the Close column.
To find the outliers, use the IQR method.
